# Import Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Salary_dataset.csv')
df.head()
```

```python
plt.scatter(x="YearsExperience",y="Salary",data=df)
plt.xlabel("Years Of Experience")
plt.ylabel("Salary")
plt.title("Years vs Salary")
```

```python
# For Test size of 20 percent
X1 = df.iloc[:, :-1].values
y1 = df.iloc[:, 1].values

# For Test size of 25 percent
X2 = df.iloc[:, :-1].values
y2 = df.iloc[:, 1].values

# For Test size of 30 percent
X3 = df.iloc[:, :-1].values
y3 = df.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split

# For test 1:
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=10)

# For test 2:
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.10, random_state=10)

# For test 3:
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.15, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression

regressor1 = LinearRegression()
regressor1.fit(X_train1, y_train1)

regressor2 = LinearRegression()
regressor2.fit(X_train2, y_train2)

regressor3 = LinearRegression()
regressor3.fit(X_train3, y_train3)
```

```python
y_pred1 = regressor1.predict(X_test1)
x_pred1 = regressor1.predict(X_train1)

y_pred2 = regressor2.predict(X_test2)
x_pred2 = regressor2.predict(X_train2)

y_pred3 = regressor3.predict(X_test3)
x_pred3 = regressor3.predict(X_train3)
```

```python
plt.scatter(X_train1, y_train1, color='red')
plt.plot(X_train1, x_pred1, color='blue')
plt.title('Linear Regression for Test Size of 20%')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X_train2, y_train2, color='red')
plt.plot(X_train2, x_pred2, color='blue')
plt.title('Linear Regression for Test Size of 10%')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X_train3, y_train3, color='red')
plt.plot(X_train3, x_pred3, color='blue')
plt.title('Linear Regression for Test Size of 15%')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
a = np.array(1.5).reshape(1, -1)
print(int(regressor1.predict(a)))
print(int(regressor2.predict(a)))
print(int(regressor3.predict(a)))
```

```python
from sklearn.metrics import mean_squared_error
rms1 = mean_squared_error(y_pred1,y_test1, squared=False)
rms2 = mean_squared_error(y_pred2,y_test2, squared=False)
rms3 = mean_squared_error(y_pred3,y_test3, squared=False)
print(rms1)
print(rms2)
print(rms3)
```

```python
print('Model Accuracy for 20% split: ', regressor1.score(X_test1, y_test1))
print('Model Accuracy for 15% split: ', regressor3.score(X_test3, y_test3))
print('Model Accuracy for 10% split: ', regressor2.score(X_test2, y_test2))
```
